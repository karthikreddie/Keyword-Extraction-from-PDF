# Extract Keywords From PDF Document 

### PyPDF2 is one such library that is used to read text from the pdf document

```python
import PyPDF2
```

##### Open the pdf file and store it in a variable to be used next

```python
filename = 'JavaBasics-notes.pdf' 
```

```python
PdfFile = open(filename,'rb') 
```

##### Read the pdf file

```python
ReadPdf = PyPDF2.PdfFileReader(PdfFile)
```

##### Extract the text from the pdf file

```python
num_pages = ReadPdf.numPages
count = 0
text = ""

while count < num_pages:
    pageObj = ReadPdf.getPage(count)
    count += 1
    text += pageObj.extractText()

print(text)
```

## The Natural Language Toolkit, or more commonly NLTK, is a suite of libraries and programs for symbolic and statistical natural language processing for English written in the Python programming language

##### Lets import the library nltk
##### use nltk.download() if not installed already

##### - Tokenization returns a list of the words in the string, using sep as the delimiter
##### - Stopwords are frequently occuring words such as 'the', 'of' etc . these words need to be removed to calculate weights of actual keywords

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

##### RE is Regular Expressions Library used for cleaning text document

```python
import re
clean1 = re.sub('[^ a-zA-Z]+','',text) #except words starting from a-z & A-Z remove everything else
clean2 = re.sub('([A-Z])', r' \1', clean1) #Split words if combined with a capital letter
print(clean2)
```

```python
shortword = re.compile(r'\s*\b\w{1,3}\b') #removed words that have just three letters
newtext = shortword.sub('',clean2)
print(newtext)
```

##### Easiest way to split a text with calling tokenizer

```python
tokens = newtext.split()
print(tokens)
```

#### Remove words like 'the', 'of' etc which hold no value since they exist in every document

##### Print words that partially matter for now

```python
stop_words = stopwords.words('english')
generalwords = [word for word in tokens if not word in stop_words]
print(generalwords)
```

#### Calculate word frequency

```python
from collections import Counter
c = Counter(generalwords)
for word, count in c.most_common():
    print(word,count)
```
